```python
import matplotlib.pyplot as plt

import numpy as np

import sympy as sp
```

### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
x = sp.symbols('x')

f_x = 10*x - 5
df_x = sp.diff(f_x, x)

print("Похідна функції 10x-5:", df_x)
```

```python
x = np.linspace(0, 10, 10)

f_x = 10*x - 5
df_x = 10 + x * 0

plt.figure(figsize=(10, 5))
plt.plot(x, f_x, label='f(x) = 10x-5', color='blue')
plt.plot(x, df_x, label="f'(x) = 10", color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sinx = sp.sin(x)
df_sinx = sp.diff(f_sinx, x)

f_lnx = sp.ln(x)
df_lnx = sp.diff(f_lnx, x)

print("Похідна функції sin(x):", df_sinx)
print("Похідна функції ln(x):", df_lnx)
```

```python
x = np.linspace(-2 * np.pi , 2 * np.pi , 100)

f_sinx = np.sin(x)
df_sinx = np.cos(x)

f_lnx = np.log(x)
df_lnx = 1 / x

plt.figure(figsize=(10, 5))
plt.plot(x, f_sinx, label='f(x) = sin(x)', color='blue')
plt.plot(x, df_sinx, label="f'(x) = cos(x)", color='red')

plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(0 , 10 , 100)

f_lnx = np.log(x)
df_lnx = 1 / x

plt.figure(figsize=(10, 5))
plt.plot(x, f_lnx, label='f(x) = ln(x)', color='blue')
plt.plot(x, df_lnx, label="f'(x) = 1 / x", color='red')

plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f_x = x ** 2 - 5 * x + 1
df_x = sp.diff(f_x, x)

print("Похідна функції x^2 - 5x + 1:", df_x)
```

```python
x = np.linspace(0 , 100 , 100)

f_x = x ** 2 - 5 * x + 1
df_x = 2 * x - 5

plt.figure(figsize=(10, 5))
plt.plot(x, f_x, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df_x, label="f'(x) = 2x - 5", color='red')

plt.title("Графік функції x^2 - 5x + 1 та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
